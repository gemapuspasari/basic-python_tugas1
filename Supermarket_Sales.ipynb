**Supermarket Sales**

Data penjualan dari 3 cabang sebuah perusahaan supermarket selama 3 bulan.

Metode prediksi data analisis dapat diaplikasi pada dataset ini

1.   Invoice ID : slip pembayaran
2.   Branch : cabang
3.   City : kota tempat supermarket berada
4.   Costumer type : tipe pembeli berdasarkan kepemilikan kartu anggota
5.   Gender : jenis kelamin pembeli
6.   Product line : Jenis produk berupa terdiri dari aksesoris elektronik, aksesoris fashion, makanan dan minuman, kesehatan dan kecantikan, keperluan rumah dan gaya hidup, olahraga dan perjalanan
7.   Unit price : harga barang
8.   Quantity : Jumlah barang yang dibeli pembeli
9.    Tax : 5% pajak yang dibayarkan konsumen
10.   Date : pembelian (Januari-Maret 2019)
11.   Time : waktu pembelian (10 am - 9 pm)
12.   Payment : metode pembayaran (tunai, kredit, Ewallet)
13.   COGS : Cost of Goods Sold (harga dasar barang yang dijual)
14.   Rating : Nilai yang diberikan pembeli atas kepuasan pengalaman belanja

Source: https://www.kaggle.com/aungpyaeap/supermarket-sales

```
import pandas as pd
```

```
database = pd.read_csv('/content/drive/MyDrive/supermarket_sales - Sheet1.csv')
database.head()
```

Menghilangkan nilai Nan atau null dalam database

```
# menghilangkan nilai NaN atau null dalam database
database_no_null = database.dropna()
database_no_null
```

Mengurutkan data berdasarkan tanggal pembelian dari Januari hingga Maret 2019

```
# mengurutkan data berdasarkan tanggal kejadian dari pembelian terbaru
database_sorted_recency = database.sort_values(['Date'], ascending = False)
database_sorted_recency
```

Mengurutkan data berdasarkan tanggal pembelian dari Januari hingga Maret 2019

```
# mengurutkan data berdasarkan tanggal kejadian dari pembelian terlama
database_sorted_recency = database.sort_values(['Date'], ascending = True)
database_sorted_recency
```

Mengurutkan data berdasarkan total pembelian (dari besar ke kecil)

```
# mengurutkan data berdasarkan total pembelian dari besar ke kecil
database_sorted_total = database_no_null.sort_values(['Total'], ascending = False)
database_sorted_total
```

Jenis produk yang paling sering dibeli

```
# jenis produk dan jumlahnya
database.groupby("Product line")["Quantity"].count()
```
